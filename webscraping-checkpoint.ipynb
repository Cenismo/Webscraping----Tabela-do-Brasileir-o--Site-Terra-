```python
import html5lib
import requests
import pandas as pd
from bs4 import BeautifulSoup

req = requests.get('https://www.terra.com.br/esportes/futebol/brasileiro-serie-a/tabela/')

if req.status_code == 200:
    print('Página disponível')
    conteudo = req.content
    soup = BeautifulSoup(conteudo, 'html.parser')
    tabela = soup.find("article",{"class":"freecontent__content", "itemprop":"articleBody"})
    tabela_str = str(tabela)
    df = pd.read_html(tabela_str)
    display(df)
```

```python
dfs = df[0]
# Extraindo colunas                                                                                                          
df2 = dfs[['Times','Times.2','P', 'J','V','E','D','GP','GC','SG','%', 'Times.3']]
print(df2)
```

```python
df2
```

```python
#removendo o ultimo caracter da ultima coluna, pois não quero manter a qtd de posições que o time se movimentou em relação a odada passada

df2['Movimentacao'] = df2['Times.3'].str[:-1]

df2.drop('Times.3', axis=1, inplace=True)
df2
```

```python
#Substituindo valores NaN por "Manteve", pois significa que o time manteve a posição em  relação a rodada passada
df2["Movimentacao"].fillna("Manteve", inplace=True)
df2
```

```python
# Renomeando colunas
df2.rename(columns={'Times':'Colocação','Times.2':'Clube'},inplace=True)
print(df2.columns)
```

```python
df2
```

```python
# retirar o ">>" nos nomes dos times
df2['Clube'] = df2['Clube'].str[:-2]
```

```python
df2
```

```python
# Obtendo número de gols por jogo
df2["Gols por jogo"] = (df2["GP"]/df2["J"])

# Obtendo número de gols tomados por jogo
df2["Gols tomados por jogo"] = (df2["GC"]/df2["J"])

# Obtendo % de vitórias
df2["% de vitórias"] = (df2["V"]/df2["J"]) *100

# Obtendo % de empates
df2["% de empates"] = (df2["E"]/df2["J"]) *100

# Obtendo % de derrotas
df2["% de derrotas"] = (df2["D"]/df2["J"]) *100
```

```python
df2
```

```python
# Salvando tabela para excel e removendo coluna index

df2.to_excel('tabelabrasileirao.xlsx', index=False)
```
